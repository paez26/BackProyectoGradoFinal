```python
import pandas as pd
from pydriller import Repository
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set() 
from gitclone import main
from joblib import load
```

```python
df = pd.DataFrame()

repo = Repository("repos/Group22-MobileApp-Grupo22-Kotlin.git", only_in_branch="Dev")
repo_commits = repo.traverse_commits()
for commit in repo_commits:
        df = pd.concat([df, pd.DataFrame({
                                        "msg" : [commit.msg],
                                                'author_name': [commit.author.name],
                                                'commiter_email': [commit.committer.email],
                                                'dmm_unit_size ': [commit.dmm_unit_size],
                                                'dmm_unit_complexity ': [commit.dmm_unit_complexity],
                                                'dmm_unit_interfacing ': [commit.dmm_unit_interfacing]
                                                })])
df = df[~df['commiter_email'].str.contains('noreply@github.com')]
# delete the rows that columns msg has merge in it
df = df[~df['msg'].str.contains("merge")]
df = df[~df['msg'].str.contains("Merge")]
df['commiter_email'] = df['commiter_email'].str.split('@').str[0]
df['commiter_email'] = df['commiter_email'].str.split('+').str[-1]
df = df.drop(columns=['author_name'])
df = df.rename(columns={'commiter_email': 'author_name'})
df = df[~df['msg'].str.contains('Lisan Algaib')]
# Clean column names
df.columns = df.columns.str.strip()

# Filter the DataFrame by author_name
filtered_df = df[df['author_name'] == 'montejs3']

# Calculate the mean of dmm_unit_size
promedio_dmm_unit_size = filtered_df['dmm_unit_size'].mean()

# Calculate the mean of dmm_unit_complexity
promedio_dmm_unit_complexity = filtered_df['dmm_unit_complexity'].mean()

# Calculate the mean of dmm_unit_interfacing
promedio_dmm_unit_interfacing = filtered_df['dmm_unit_interfacing'].mean()

print("Promedio DMM Unit Size:", promedio_dmm_unit_size)
print("Promedio DMM Unit Complexity:", promedio_dmm_unit_complexity)
print("Promedio DMM Unit Interfacing:", promedio_dmm_unit_interfacing)


```
